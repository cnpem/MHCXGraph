```python
crossDict = {}

for i in range(37):
    crossDict[i] = {"sim": set(), "não": set()}

for i in range(37):
    TCR_1 = set(crossreact[crossreact["peptide_id"] == i]["TCR_pair_id"].tolist())

    for j in range(i + 1, 37):
        TCR_2 = set(crossreact[crossreact["peptide_id"] == j]["TCR_pair_id"].tolist())

        cross = TCR_1.intersection(TCR_2)

        if len(cross) == 0:
            crossDict[i]["não"].add(j)
            crossDict[j]["não"].add(i)
        else:
            crossDict[i]["sim"].add(j)
            crossDict[j]["sim"].add(i)
```

```python
groups = {}

for i in crossDict.keys():
    peps = crossDict[i]["não"]

    if not peps:
        groups[i] = set()
        continue

    intersection = set.intersection(*(crossDict[pep]["não"] for pep in peps))

    # remove o próprio peptide se aparecer
    intersection.discard(i)

    groups[i] = intersection
```

```python
groupsUnique = []
for i in groups.keys():
    groups[i].add(i)
    if tuple(groups[i]) not in groupsUnique:
        if len(groups[i]) > 1:
            groupsUnique.append(tuple(groups[i]))
```
